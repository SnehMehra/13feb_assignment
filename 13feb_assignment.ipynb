ZeroDivisionError:
This exception is raised when a division or modulo operation is performed with zero as the divisor.
Example:



```python
dividend = 10
divisor = 0

try:
    result = dividend / divisor
except ZeroDivisionError as e:
    print("Error:", e)
```

OverflowError:
This exception is raised when the result of an arithmetic operation is too large to be represented within the available numeric range.
Example:

```python
x = 10 ** 1000

try:
    result = x * x
except OverflowError as e:
    print("Error:", e)

```

Q.4

The LookupError class is used as a base class for exceptions that occur when a lookup or indexing operation fails

KeyError:
This exception is raised when a dictionary or mapping structure is accessed using a key that does not exist in the structure.
Example:

```python
my_dict = {"apple": 1, "banana": 2, "orange": 3}

try:
    value = my_dict["grape"]
except KeyError as e:
    print("Error:", e)
```

IndexError:
This exception is raised when attempting to access a sequence (e.g., a list or string) using an index that is out of range.


```python
my_list = [1, 2, 3]

try:
    value = my_list[5]
except IndexError as e:
    print("Error:", e)
```

ImportError:
ImportError is a base class for exceptions raised when importing a module or a specific object from a module encounters an error. It can occur due to various reasons, such as the module or object not being found, circular imports, or issues with module dependencies
